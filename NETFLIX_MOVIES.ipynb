CONVERTING DATA TYPE OF OBJECT INTO DATE TYPE

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])

print(df['Release_Date'].dtypes)
```

EXTRACTING ONLY YEAR IN THE DATE FORMATE

```python
df['Release_Date'] = df['Release_Date'].dt.year
```

```python
df.head()
```

DELETING THE COLUMNS THAT ARE NOT USEFUL FOR THE ANALYSIS

```python
cols = ['Overview','Original_Language','Poster_Url']

df.drop(cols,axis=1,inplace=True)
```

```python
df.head()
```

DELETING A NAN VALUES IN COLUMNS

```python
df = df.dropna()

df.isna().sum()
```

```python
df['Popularity'].isna().sum()
```

```python
def Categorized_Value(value):
    if value >= 7.0:
        return "Popular"
    elif value >= 6.0:
        return "Average"
    elif value >= 5.0:
        return "Not Average"
    else:
        return "Not Popular"

df['Vote_Average'] = df['Vote_Average'].apply(Categorized_Value)
```

```python
df.head(20)
```

```python
df.tail(20)
```

```python
df['Vote_Average'].value_counts()
```

SPLITING AND EXPLODED GENRE COLUMN

```python
df['Genre'] = df['Genre'].str.split(',')

df = df.explode('Genre').reset_index(drop=True)
```

```python
df.head(10)
```

BUSINESS PROBLEM

1.What is the most frequent genre of movies release on Netflix?

```python
df['Genre'].describe()
```

```python
plt.figure(figsize=(8, 8))

sns.countplot(
    y='Genre',
    data=df,
    order=df['Genre'].value_counts().index,
    color='Green'
)

plt.title("Genre Column Distribution")
plt.tight_layout()
plt.show()
```

2.WHICH HAS HIGHEST VOTE IN VOTE_AVERAGE COLUMN?

```python

sns.catplot(
    y='Vote_Average',
    data=df,
    kind = 'count',
    order=df['Vote_Average'].value_counts().index,
    color='Red'
)

plt.title("Vote_Average Column Distribution")
plt.show()
```

WHICH MOVIE GOT THE HIGHEST POPULARITY ?WHAT ITS GENRE?

```python
df[df['Popularity'] == df['Popularity'].max()]
```

WHICH MOVIE GOT THE LOWEST POPULARITY ?WHAT ITS GENRE?
